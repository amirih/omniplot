```python
from omniplot  import plot as op
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.print_figure_kwargs={'bbox_inches': None}
#%matplotlib tk
```

```python
df=sns.load_dataset("penguins")
df=df.dropna(axis=0)
print(df.shape)
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species'],
                save="scatter1.png")
```

```python
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species'],
                sizes="body_mass_g",
                save="scatter2.png")
```

```python
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species', "sex"],
                sizes="body_mass_g",
                save="scatter3.png")
```

```python
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species', "sex", "island"],
               colors="flipper_length_mm",
                sizes="body_mass_g",
                save="scatter4.png")
```

```python
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species', "sex"],
                sizes="body_mass_g",
                marginal_dist=True,
                save="scatter5.png", figsize=[9,4])
```

```python
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species', "sex"],
                sizes="body_mass_g",
                marginal_dist=True,
                kde=True,
                save="scatter6.png", figsize=[9,4])
```

```python
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species', "sex"],
                sizes="body_mass_g",
                marginal_dist=True,
                kde=True,
                kmeans=True,
                save="scatter7.png")
```

```python
res=op.scatterplot(df, 
               x="bill_length_mm",
               y="bill_depth_mm", 
               category=['species', "sex"],
                sizes="body_mass_g",
                marginal_dist=True,
                kde=True,
                regression=True,
                save="scatter8.png")
```
