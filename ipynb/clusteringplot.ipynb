```python
import seaborn as sns
from omniplot import plot as op
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("penguins")
df=df.dropna(axis=0)
df.shape
```

```python
df=df.sample(n=50, random_state=1)
```

```python
features=["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
dfnew=op.clusterplot(df,variables=features,category=["species","sex", "body_mass_g"],method="kmeans",n_clusters="auto")
```

```python
dfnew=op.clusterplot(df,category=["species","sex", "island"],method="kmeans",n_clusters=3,size=100)
plt.savefig("kmeans.pdf")
```

```python
#Calculate z-scores for each kmeans cluster
import numpy as np
val=df[features].values
print(val)
mean=np.mean(val,axis=0)
std=np.std(val,axis=0)
zscores=[]
for i in range(3):
    _val=val[dfnew["data"][3]["kmeans"]==i]
    _val=np.mean(_val,axis=0)
    zscores.append((_val-mean)/std)
```

```python

fig, axes=plt.subplots(1,3,figsize=(10,3))
for i in range(3):

    axes[i].bar(np.arange(0, 4),zscores[i])
    axes[i].set_title("Cluster "+str(i))
    axes[i].set_xticks(np.arange(0, 4))
    axes[i].set_xticklabels(["variable1", "variable2","variable3","variable4"],rotation=90)
    axes[i].set_ylim([-1.5,1.5])
axes[0].set_ylabel("Deviation from overall mean")
plt.savefig("zscore.pdf",dpi=300,bbox_inches="tight")
```
