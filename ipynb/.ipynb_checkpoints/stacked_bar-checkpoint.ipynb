```python
import seaborn as sns
from omniplot import plot as op
```

```python
df=sns.load_dataset("penguins")
df=df.dropna(axis=0)
res=op.stacked_barplot(df, x="species",hue="island", scale="absolute", test_pairs=[["Adelie","Gentoo"]])
```
