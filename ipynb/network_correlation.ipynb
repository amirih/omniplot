```python
import pandas as pd
from omniplot import networkplot as netp
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")
df = df.dropna(axis=0)
df = df.reset_index()
res = netp.correlation(
    df,
    category=["species", "island", "sex"],
    method="pearson",
    ztransform=True,
    clustering="louvain",
    show_edges=True,
    bundle=False,
)
```
