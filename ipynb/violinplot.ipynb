```python
import seaborn as sns
from omniplot import plot as op
```

```python
df = sns.load_dataset("penguins")
df = df.dropna(axis=0)
op.violinplot(
    df,
    x="species",
    y="bill_length_mm",
    pairs=[["Adelie", "Chinstrap"], ["Gentoo", "Chinstrap"], ["Adelie", "Gentoo"]],
    test="mannwhitneyu",
    significance="symbol",
    swarm=True,
)
```
