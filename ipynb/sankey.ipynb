```python
import pandas as pd
from omniplot import networkplot as netp
```

```python
# the below compares two different k mean cluster results and a gender category of penguins.
# you can see the k mean cluster with 3 clusters devide Adelie and Chinstrap species.
# blocks are colored based on species names.
df=pd.read_csv("../data/kmeans_result.csv")
netp.sankey_category(df, ["kmeans2","kmeans3","sex"],colormode="alternative",altcat="species")
```

```python
# the same class names are colored with a common color 
netp.sankey_category(df, ["kmeans2","kmeans3","sex"],colormode="shared",altcat="species")
```

```python
# bocks are independently colored by their categorical labels
netp.sankey_category(df, ["kmeans2","kmeans3","sex"],colormode="independent",altcat="species")
```

```python
# blcoks are colored based only on the most left bock colors 
# so that you can see how the samples with the original labels 
# are classified in the different categorical labels 
netp.sankey_category(df, ["kmeans2","kmeans3","sex"],colormode="trace")
```
