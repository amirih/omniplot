```python
import seaborn as sns
from omniplot import plot as op
```

```python
df=sns.load_dataset("penguins")
df=df.dropna(axis=0)
tmp=[]
for num, (sp, i ,se) in enumerate(zip(df["species"], df["island"],df["sex"])):
    if num/df.shape[0] > 0.98:
        tmp.append("NA")
    else:
        tmp.append(sp[0]+"_"+i[0]+"_"+se[0])
df["combine"]=tmp
#print(df)
res=op.nice_piechart(df, category=["species", "island","sex","species", "island","sex","combine"],ncols=4)
```

```python
res=op.nice_piechart(df, category=["species", "island","sex","species", "island","sex","combine"],ncols=3, hatch=True)
```
